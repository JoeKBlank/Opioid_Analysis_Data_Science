This notebook is being used to plot multidimensional data and visuals multidimensional regression models between our predictors

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
tots_2017 = pd.read_csv("totals2017.csv")
tots_2018 = pd.read_csv("totals2018.csv")
```

```python
x_max = tots_2017['DepressionValue'].max()
y_max = tots_2017['EducationLTHSValue'].max()

x_min = tots_2017['DepressionValue'].min()
y_min = tots_2017['EducationLTHSValue'].min()

x_range = 30
y_range = 30
```

```python
X = tots_2017[['DepressionValue','EducationLTHSValue']].values.reshape(-1,2) #X and Z values
Y = tots_2017['Normalized_Death_Rate'] #Y values
```

```python
x = X[:, 0]
y = X[:, 1]
z = Y
```

```python
x_pred = np.linspace(x_min, x_max, x_range)   
y_pred = np.linspace(y_min, y_max, y_range) 

xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)
model_viz = np.array([xx_pred.flatten(), yy_pred.flatten()]).T
```

```python
lr = linear_model.LinearRegression()

model = lr.fit(X, Y)
predicted = model.predict(model_viz)
#predicted = model.predict(Y2)

#r2 = model.score(X2, Y2)
#R2 = r2_score(Y2,predicted)
#BayesianRidge()
#LinearRegression()
```

```python
plt.style.use('default')

fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')

axes = [ax1, ax2, ax3]
count = 0;
for ax in axes:

    ax.plot(x, y, z, color='k', zorder=15, linestyle='none', marker='o', alpha=0.5) #alpha was 0.5
    ax.scatter(xx_pred.flatten(), yy_pred.flatten(), predicted, facecolor=(0,0,0,0), s=20, edgecolor='#70b3f0')
    ax.set_xlabel('Depression', fontsize=12)
    ax.set_ylabel('Education', fontsize=12)
    ax.set_zlabel('Death Rates', fontsize=12)
    ax.locator_params(nbins=4, axis='x')
    ax.locator_params(nbins=5, axis='x')
    count += 1

#ax1.text2D(0.2, 0.32,'CS 6010', fontsize=13, ha='center', va='center',
           #transform=ax1.transAxes, color='grey', alpha=0.5)
#ax2.text2D(0.3, 0.42, 'CS 6010', fontsize=13, ha='center', va='center',
           #transform=ax2.transAxes, color='grey', alpha=0.5)
#ax3.text2D(0.85, 0.85, 'CS 6010', fontsize=13, ha='center', va='center',
           #transform=ax3.transAxes, color='grey', alpha=0.5)

ax1.view_init(elev=28, azim=120)
ax2.view_init(elev=4, azim=114)
ax3.view_init(elev=60, azim=165)

fig.suptitle('2017 Education and Depression vs Death Rates', fontsize=20)

fig.tight_layout()
```
